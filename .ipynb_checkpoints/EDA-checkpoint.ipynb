## Pokemon Data Cleaning and Exploratory Data Analysis
Data sources:

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print('setup complete')
````

````python
type_list = ['fire', 'water', 'grass', 'fighting', 'psychic', 'ghost', 'dark',
            'dragon', 'fairy', 'steel', 'ground', 'rock', 'flying', 'electric',
            'poison', 'bug', 'normal', 'ice']
stat_list = ['hp', 'atk', 'physdef', 'spatk', 'spdef', 'speed']
stat_list_bst = ['hp', 'atk', 'physdef', 'spatk', 'spdef', 'speed', 'bst']
````

````python
#resetting the datatypes of the csv to conserve space and run faster

def set_datatypes(df):
    df['type1'] = df['type1'].astype('category')
    df['type2'] = df['type2'].astype('category')
    df[stat_list] = df[stat_list].astype('int16')
    df['alternate'] = df['alternate'].astype('boolean')
    df['tier'] = df['tier'].astype('category')
    return df
````

````python
# adding a "bst" column that represents the sum of all 6 stats

def add_bst(df):
    df['bst'] = (df['hp'] + df['atk'] + df['physdef'] + df['spatk'] \
        + df['spdef'] + df['speed']).astype('int16')
    return df
````

````python
temp_df = pd.read_csv("Natdex_Data.csv")
temp_df.to_csv("Natdex_Data_copy.csv")
````

````python
# read in the data and apply the above functions

pokemon_df = pd.read_csv("Natdex_Data.csv")

pokemon_df = set_datatypes(pokemon_df)
pokemon_df = add_bst(pokemon_df)

pokemon_df
````

````python
# adding methods to query for type and ability in both columns. Not necessary but will make code easier to read


from pandas.core.base import PandasObject


def query_type(self, thetype):
    '''
    uses the df.query function to query for all the rows of the dataframe where either the 'type1' or 'type2' columns are the given type.
    
    Parameters:
        self (Pandas DataFrame): must have type1 and type2 columns (always true with my use cases)
        thetype (str): the type you are tryig to query for
    
    Returns:
        Pandas DataFrame: a subset of the input dataframe where type1 or type2 is equal to 'thetype'
    '''
    
    return self.query("type1 == @thetype or type2 == @thetype")



def query_ability(self, ability):
    '''
    uses the df.query function to query for all the rows of the dataframe where 'ability1', 'ability2', or 'hiddenability3'are equal to the given ability.
    
    Parameters:
        self (Pandas DataFrame): must have ability1, ability2, and hiddenability columns (always true with my use case)
        ability (str): the ability you are trying to query for
    
    Returns:
        Pandas DataFrame: a subset of the input dataframe where ability1, ability2, or hiddenability is equal to 'ability'
    '''
    
    return self.query("ability1 == @ability or ability2 == @ability or hiddenability == @ability")



def has_type(self, thetype):
    '''
    Determines whether the input contains a given type in its 'type1' or 'type2' columns. Returns True or False
    
    Parameters:
        self (Pandas DataFrame): must have 'type1' and 'type2' columns (always true in my use cases). Functionality is built for DataFrames of different lengths
        thetype (str): the type that you are checking for. Make sure that it is a valid type because if you have a typo or mistake, the method will always return False
    
    Returns:
        boolean: True if the input contains the type, False otherwise
    '''
    # check if DataFrame is appropriate length
    if (len(self) <= 0):
        raise Exception('length of input dataframe is invalid (<= 0)')
    
    #check if DataFrame is length 1 (no need to iterate if it is)
    elif (len(self) == 1):
        return (self['type1'].values[0] == thetype) or (self['type2'].values[0] == thetype)
    
    # iterate to determine whether it has the type or not
    else:
        for i, pokemon in self.iterrows():
            if ((pokemon['type1'] == thetype) or (pokemon['type2'] == thetype)):
                return True
        return False
    
    
    
def has_ability(self, ability):
    '''
    Determines whether the input contains a given ability in its 'ability1', 'ability2', or 'hiddenability' columns. Returns True or False
    
    Parameters:
        self (Pandas DataFrame): must have 'type1' and 'type2' columns (always true in my use cases). Functionality is built for DataFrames of different lengths
        ability (str): the ability that you are checking for. Make sure that it is a valid ability because if you have a typo or mistake, the method will always return False
    
    Returns:
        boolean: True if the input contains the ability, False otherwise
    '''
    #check if DataFrame is appropriate length
    if (len(self) <= 0):
        raise Exception('length of input dataframe is invalid (<=0)')
    # check if DataFrame is length 1 (no need to iterate if it is)
    elif (len(self) == 1):
        return (self['ability1'].values[0] == ability or self['ability2'].values[0] == ability or 
            self['hiddenability'].values[0] == ability)
        
    #iterate to determine whether it has the ability or not
    else:
        mylist = []
        for i, pokemon in self.itterrows:
            if (pokemon['ability1'] == ability or pokemon['ability2'] == ability or 
                          pokemon['hiddenability'] == ability):
                return True
        return False
            
PandasObject.query_type = query_type
PandasObject.query_ability = query_ability
PandasObject.has_type = has_type
PandasObject.has_ability = has_ability
````

Here's the reason why there are so many pokemon in RU. Source: my experience

The smogon tiering system typically has 5 standard tiers: OU, UU, RU, NU, PU, descending in that order. Each tier has better pokemon than the ones below it. In Natdex, however, there isn't enough players to justify an NU or a PU. Therefore, every pokemon that would be NU or PU or Untiered by usage is just lumped in in RU.

I'll explain the UUBL, Uber, and AG tiers. First: Uber. Ubers is where all the banned pokemon from OU go. For my purposes, I'll count this as viable. Above Uber is AG (Anything Goes), where the banned pokemon from Uber go. That will also be considered viable. Finally, UUBL is where all the banned pokemon from UU go. Too strong for UU, not good enough for OU. This will be counted as not viable.

Nonetheless, this isn't important for my purposes. I'm going to be conducting this OVA style. I'll add that in now.

````python
def create_viability(tier: str):
    '''
    When applied to pokemon_df['tier'], will return a column that represents whether the pokemon is good in OU or not.
    
    Parameters:
        tier (str): must be one of the recognized tiers. Exception will be thrown otherwise
        
    Returns:
        bool: true for viable, false for not
    '''
    viable_list = ['OU', 'Uber', 'AG']
    # extra entries are added to unviable_list to future proof against tiers that currently have no pokemon
    unviable_list = ['UUBL', 'UU', 'RUBL', 'RU', 'NFE', 'LC', 'LC Uber', '(OU)', '(UU)', '(RU)']
    
    if (tier in viable_list):
        return True
    elif (tier in unviable_list):
        return False
    else:
        raise Exception("Unidentified Tier")
````

````python
pokemon_df['isViable'] = pokemon_df['tier'].apply(create_viability)
````

### Check for relationship between types and tier/viability

I'm creating a dataframe where the index is the type and the column is the proportion of pokemon with that type that are either viable or not. I've decided to use proportion to account for the differing number of pokemon in each type

````python
# viewing the proportion of the viable pokemon with each type

total_list = []
for thetype in type_list:
    
    #get the total number of pokemon with that type
    num_type = evolved_df.query_type(thetype)['name'].count()
    
    #get the proportion of viable pokemon with that type
    num_viable = evolved_df.query_type(thetype).query('isViable == True')['name'].count()
    num_viable /= num_type
    
    
    #get the amount of unviable pokemon with that type
    num_unviable = evolved_df.query_type(thetype).query('isViable == False')['name'].count()
    num_unviable /= num_type
    
    #create a series and append it to the list
    tempSeries = pd.Series([num_viable, num_unviable], index=['Viable', 'Unviable'], name=thetype)
    total_list.append(tempSeries)
    
type_viability = pd.DataFrame(total_list)
````

Let's check if Legendary status plays a role with the chi squared test for independence

````python
legend_crosstab = pd.crosstab(evolved_df['isViable'], evolved_df['isLegend'])
legend_crosstab
````

Assumptions:
* Variables are independent: assumed
* Large counts: all expected values are above 5 (lowest is ~18.5)

Hypotheses:
* Ho: isLegend and isViable are independent
* Ha: isLegend and isViable are not independent

````python
from scipy.stats import chi2_contingency

legend_chi2_result = chi2_contingency(legend_crosstab)

print("Statistic:", legend_chi2_result.statistic, "\nP-value:", legend_chi2_result.pvalue)
````

The statistic is greater than the P-value by many, many orders of magnitude, therefore, reject the null hypothesis

We can conclude that the legendary status and the viability of a pokemon are not independent

````python
print(legend_chi2_result.expected_freq)
legend_crosstab
````

Looking at the expected values versus the actual, it's clear that more legendary pokemon are viable than expected, and less non-legendary pokemon are unviable than expected. Therefore, if you're a legendary pokemon, you're more likely to be viable

Let's do the same for alternate forms

````python
alt_crosstab = pd.crosstab(evolved_df['isViable'], evolved_df['alternate'])
alt_crosstab
````

Assumptions:
* The two categories are independent: assumed
* Large counts: all expected values are above 5 (lowest is ~25)

Hypotheses:
* Ho: alternate and isLegend are independent
* Ha: alternate and isLegend are dependent

````python
alt_chi2_result = chi2_contingency(alt_crosstab)

print("Statistic:", alt_chi2_result.statistic, "\nP-value:", alt_chi2_result.pvalue)
````

The statistic is significantly greater than the p-value, therefore reject the null hypothesis

We can conclude that alternate and isLegend are not independent

````python
print(alt_chi2_result.expected_freq)
alt_crosstab
````

Looking at the expected values versus the actual, it's clear that it's the same case with alternates. More alternate formed pokemon are viable than expected, to a significant degree. This is likely because a lot of legendaries have alternate forms, as well as the megas, which are marked as alternate forms and have a higher bst. This also includes the arceus forms
